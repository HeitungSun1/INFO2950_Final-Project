**Data analysis**

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
happiness = pd.read_csv("happiness.csv")
education_pretrim = pd.read_csv("education.csv")
education_trim = education_pretrim.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
combined = pd.merge(happiness, education_trim, on='Country', how='inner')
print("Dataset Mean Values:\n",combined.mean(), "\n")
print("Dataset Standard Deviation Values:\n", combined.std(),"\n")
```

```python
plt.scatter(combined["Happiness Score"], combined["Education Index"])
plt.title("Scatter Plot of Happiness Score vs Education Index")
plt.xlabel("Happiness Score")
plt.ylabel("Education Index")
plt.show()
```

```python
plt.scatter(combined["Happiness Score"], combined["Expected Years of Schooling"])
plt.title("Scatter Plot of Happiness Score vs Expected Years of Schooling")
plt.xlabel("Happiness Score")
plt.ylabel("Expected Years of Schooling")
plt.show()
```

```python
plt.scatter(combined["Happiness Score"], combined["Economy (GDP per Capita)"])
plt.title("Scatter Plot of Happiness Score vs Economy (GDP per Capita)")
plt.xlabel("Happiness Score")
plt.ylabel("Economy (GDP per Capita)")
plt.show()
```

**Evaluation of significance**

```python
combined = pd.merge(happiness, education_trim, on='Country', how='inner')
combined
```

```python
x = combined['Happiness Score']
y = combined['Education Index']
plt.xlabel('Happiness Score')
plt.ylabel('Education Index')
plt.title("Scatter Plot of Happiness Score vs Education Index")
plt.xlim([2, 9]);
plt.ylim([0, 1]);
plt.scatter(x, y)
plt.show()
plt.scatter(y, x)
plt.xlabel('Education Index')
plt.ylabel('Happiness Score')
plt.title("Scatter Plot of Education Index vs Happiness Score")
plt.show()
print("As Happiness Score increases, Education Index seems to increase. The variables appear to be positively correlated.")

```

```python
x = combined['Happiness Score']
y = combined['Education Index']
n = len(x)

mean_x = x.mean()
mean_y = y.mean()

std_x = x.std()
std_y = y.std()

print("Var(Happiness Score):", numpy.sum((x - mean_x)**2) / n, x.var())
print("Var(Education Index):", numpy.sum((y - mean_y)**2) / n, y.var())
print("cov(Happiness Score, Education Index) dividing by n:  ", numpy.sum((x - mean_x) * (y - mean_y)) / n, numpy.cov(x, y, ddof=0)[0,1])
print("cov(Happiness Score, Education Index) dividing by n-1:", numpy.sum((x - mean_x) * (y - mean_y)) / (n-1),
  	numpy.cov(x, y)[0,1])
print("corr(Happiness Score, Education Index):", np.corrcoef(combined['Happiness Score'], combined['Education Index']))
```

```python
from sklearn.linear_model import LinearRegression
ha = combined['Happiness Score'].values.reshape(-1,1)
ed = combined['Education Index'].values.reshape(-1,1)
model = LinearRegression().fit(ha, ed)
model.intercept_[0]
model.coef_[0][0]
print("intercept of Happiness Score and Education Index: ",model.intercept_[0])
print("slope of Happiness Score and Education Index: ",model.coef_[0][0])
```

**Conclusion**

The finding of consistently linear patterns of education level and hapiness score across different countries substantiates the possibility that the education level and happiness score are positively correlated. The results demonstrate the positive relationship between both eduaction level and happiness and GDP and happiness. One special problem we noticed during the data analysis is the faireness of the raw data. Since happiness level is a relatively subject evaluation and it's hard to standardize, the results in terms of "happiness score" might be biased. For example, two people even have exactly the same things and with same background they may have different "happiness score" due to their different definition of "happiness." To block these potential problems in raw data, we assume that "happiness score" is authentic enough to do the data analysis. In this way, the estimate of the relationship between education level and happiness, and how education level affect happiness comparing to other factors, are thus relatively conservative.
